## Data Cleaning

```python
# filtering out those rows whose status is either Stump or Dead
mask = ((tree_census_subset["status"] == "Stump") | (tree_census_subset["status"] == "Dead"))
```

```python
# fill all the NaN values in the masked set with "Not Applicable"
tree_census_subset.loc[mask] = tree_census_subset.loc[mask].fillna("Not Applicable")
```

```python
tree_census_subset[tree_census_subset["status"] == "Stump"]
```

```python
# check for more NaN values
tree_census_subset.isna().sum()
```

```python
# based on the above analysis, there are still some NaN values but not many of them,
# here, fill the NaN values with the most frequent values from their counterparts
tree_census_subset["tree_dbh"].fillna(4.0, inplace=True)
tree_census_subset["stump_diam"].fillna(0.0, inplace=True)
tree_census_subset["curb_loc"].fillna("OnCurb", inplace=True)
tree_census_subset["status"].fillna("Alive", inplace=True)
tree_census_subset["health"].fillna("Good", inplace=True)
tree_census_subset["spc_latin"].fillna("No Observation", inplace=True)
tree_census_subset["steward"].fillna("None", inplace=True)
tree_census_subset["sidewalk"].fillna("NoDamage", inplace=True)
tree_census_subset["problems"].fillna("None", inplace=True)
tree_census_subset["root_stone"].fillna("No", inplace=True)
tree_census_subset["root_grate"].fillna("No", inplace=True)
tree_census_subset["root_other"].fillna("No", inplace=True)
tree_census_subset["trunk_wire"].fillna("No", inplace=True)
tree_census_subset["trnk_light"].fillna("No", inplace=True)
tree_census_subset["trnk_other"].fillna("No", inplace=True)
tree_census_subset["brch_light"].fillna("No", inplace=True)
tree_census_subset["brch_shoe"].fillna("No", inplace=True)
tree_census_subset["brch_other"].fillna("No", inplace=True)
```

```python
# there are no NaN values
tree_census_subset.isna().sum()
```

```python
# classify further big trees as those with tree_dbh and stump_diam larger than 60
big_trees = tree_census_subset[(tree_census_subset["tree_dbh"]>60) | (tree_census_subset["stump_diam"]>60)]
big_trees
```

```python
# filter our subset further by excluding the big trees as their values are unrealistic hence unnecessary in the analysis
tree_census_subset = tree_census_subset[(tree_census_subset["tree_dbh"]<=60) | (tree_census_subset["stump_diam"]<=60)]
tree_census_subset
```

```python
# create two more subsets based on trees that have status as "Alive" and those that are either "Stump" or "Dead"
tree_census_subset_alive = tree_census_subset[tree_census_subset["status"] == "Alive"]
tree_census_subset_dead_or_stump = tree_census_subset[(tree_census_subset["status"] == "Dead") | (tree_census_subset["status"] == "Stump")]
```

```python
# stats information based on latin names and tree_dbh
stats_alive = tree_census_subset_alive.groupby("spc_latin")["tree_dbh"].describe().reset_index()[["spc_latin", "25%", "75%"]]
stats_alive
```

```python
# merge the stats_alive subset with the tree_census_subset_alive to the left based on latin names
tree_census_subset_alive.merge(stats_alive, on="spc_latin", how="left")
```

```python
# tree_dbh values that are smaller than the "lower quatile" and larger than  the "upper quatile" are abit unrealistic
# filter the subset in a way that the smallest tree_dbh value is equal to the lower quatile and the 
# largest tree_dbh value is equal to the upper quatile
p25 = tree_census_subset_alive["tree_dbh"].quantile(0.25)
p75 = tree_census_subset_alive["tree_dbh"].quantile(0.75)
mask = tree_census_subset_alive["tree_dbh"]<p25
tree_census_subset_alive.loc[mask, "tree_dbh"] = p25

mask = tree_census_subset_alive["tree_dbh"]>p75
tree_census_subset_alive.loc[mask, "tree_dbh"] = p75
```
